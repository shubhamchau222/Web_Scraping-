
#   What is web Scrapping ? 
Web scraping, web harvesting, or web data extraction is data scraping used for extracting data from websites. Web scraping software may access the World Wide Web directly using the Hypertext Transfer Protocol, or through a web browser. While web scraping can be done manually by a software user, the term typically refers to automated processes implemented using a bot or web crawler. It is a form of copying, in which specific data is gathered and copied from the web, typically into a central local database or spreadsheet, for later retrieval or analysis.

```
jobs  = soup.find_all('li' , class_ ="clearfix job-bx wht-shd-bx" )
count = 0
for job in jobs :
    company_name = job.find("h3" , class_ = "joblist-comp-name").text
    skills = job.find("span" , class_="srp-skills").text
    post_before = job.find("span" , class_ = 'sim-posted').text.strip()
    count+=1

    print("company_name : " , company_name)
    print("Required Skills : " , skills)
    print(post_before)
    print("-----"*40)

print(count)
```

```
# in above data the jobs posted before few days ago are could be outdated :
# hence to extract the current job posts

jobs  = soup.find_all('li' , class_ ="clearfix job-bx wht-shd-bx" )
count = 0
for job in jobs :
    post_before = job.find("span" , class_ = 'sim-posted').text.strip()
    if "few" not in post_before:
        company_name = job.find("h3" , class_ = "joblist-comp-name").text
        skills = job.find("span" , class_="srp-skills").text
        post_before = job.find("span" , class_ = 'sim-posted').text.strip()
    
        count+=1

        print("company_name : " , company_name)
        print("Required Skills : " , skills)
        print(post_before)
        print("-----"*40)

print(count)
```

# Saving the information 

```
jobs  = soup.find_all('li' , class_ ="clearfix job-bx wht-shd-bx" )

for index, job in enumerate(  jobs )  :
    post_before = job.find("span" , class_ = 'sim-posted').text.strip()
    if "few" not in post_before:
        company_name = job.find("h3" , class_ = "joblist-comp-name").text
        skills = job.find("span" , class_="srp-skills").text
        post_before = job.find("span" , class_ = 'sim-posted').text.strip()
        # adding href
        href = job.header.h2.a['href'] 

        #lets save the result into the text file:
        with open ("jobs_result.txt" , "a+") as f :
            f.write(f"company_name : {company_name}")
            f.write(f"Required Skills : { skills}")
            f.write(f'for more info :{href}')
            f.write(post_before)
        print("file saved....")

        print("company_name : " , company_name)
        print("Required Skills : " , skills)
        print('for more info -->' ,href )
        print(post_before)
        print("-----"*40)
```
